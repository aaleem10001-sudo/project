```python
#1 Student Class

class student:
    def __init__(self, name, age):
        self.name=name
        self.age=age

    def display(self):
        print("Name",self.name)
        print("Age",self.age)

s1=student("ali",22)
s1.display()
```

```python
#2 Employee Class

class employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

    def display(self):
        print("name:",self.name)
        print("salary:",self.salary)

e = employee("aleesha",50000000)
e.display()
```

```python
#3 Product Class

class product:
    def __init__(self , name,price):
        self.name=name
        self.price=price
    
    def display(self):
        print("name",self.name)
        print("price",self.price)

p=product("laptop",260000)
p.display()
```

```python
#4 car class

class car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display(self):
        print("brand", self.brand)
        print("model", self.model)


c = car("civic", "rebirth")
c.display()


```

```python
#5 Bank Account class

class bankAccount:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance

    def display(self):
        print("name",self.name)
        print("balance",self.balance)



a=bankAccount("habib",1736781638)
a.display()
```
